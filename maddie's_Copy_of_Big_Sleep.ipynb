```
!pip install big-sleep --upgrade

import time
import shutil
import torch
from tqdm.notebook import trange
from IPython.display import Image, display
from big_sleep import Imagine
```

```
# connect to google drive using pydrive

# Import PyDrive and associated libraries.
# This only needs to be done once in a notebook.
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once in a notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```
TEXT = "kudzu cathedral|a cathedral made of kudzu|romantic" #@param {type:"string"}
SAVE_EVERY = 100 #@param {type:"number"}
SAVE_PROGRESS = False #@param {type:"boolean"}
SAVE_BEST = True #@param {type:"boolean"}
LEARNING_RATE =  7e-2#@param {type:"number"}
ITERATIONS =  500#@param {type:"number"}
EPOCHS = 1 #@param {type:"number"}
SEED =  1#@param {type: "number"}
RESOLUTION = 512#@param {type:"number"}
NUM_ATTEMPTS = 5#@param {type:"number"}
TEXT_MIN = "blur|zoom"#@param {type:"string"}

num_attempts = NUM_ATTEMPTS

# can get code to use from simplified notebook for big sleep as well as 
# test/example code
```

```
terminate = False

def signal_handling(signum,frame):
    global terminate
    terminate = True

for attempt in range(num_attempts):
  print("attempt number ",attempt,'\n')

  model = Imagine(
      text = TEXT,
      save_every = SAVE_EVERY,
      lr = LEARNING_RATE,
      iterations = ITERATIONS,
      epochs = EPOCHS,
      save_progress = SAVE_PROGRESS,
      save_best = SAVE_BEST,
      seed = SEED,
      text_min = TEXT_MIN
      # resolution = RESOLUTION
  )

  model()

  #saving and displaying attempt
  filename = model.text_path + "-attempt-" + str(attempt) + ".png"
  print("filename is: ", filename)
  shutil.copy(model.text_path + ".best.png", filename)
  uploaded = drive.CreateFile({'title': filename})
  uploaded.SetContentFile(filename)
  uploaded.Upload()
  print('Uploaded file with ID {}'.format(uploaded.get('id')))
  print('File name is ', filename)
  image=Image(filename)
  display(image)

  #cleaning and trying again
  try:
      time.sleep(2)
      del model
      time.sleep(2)
      torch.cuda.empty_cache()
  except Exception:
      torch.cuda.empty_cache()
      torch.cuda.empty_cache()
  except Exception:
      torch.cuda.empty_cache()
```
